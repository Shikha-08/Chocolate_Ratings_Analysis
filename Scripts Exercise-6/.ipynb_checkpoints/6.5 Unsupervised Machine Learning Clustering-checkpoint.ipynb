### 1. Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans 
import pylab as pl 
```

```python
# Identify categorical columns
categorical_columns = df_cleaned_data.select_dtypes(include=['object']).columns
categorical_columns
```

```python
# Drop categorical columns
df_numeric = df_cleaned_data.drop(columns=categorical_columns)

df_numeric.head()
```

```python
# Apply Standardization
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

# Convert back to a DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=df_numeric.columns)

df_scaled.head()
```

### 2. The elbow technique

```python
# Define range for k
num_clusters = range(1, 10)

# Compute k-means inertia for each k
scores = [KMeans(n_clusters=k, random_state=42).fit(df_scaled).inertia_ for k in num_clusters]
```

```python
# Plot the elbow curve

plt.plot(num_clusters, scores, color='b')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

The curve steeply decreases from K=1 to K=3 or 4, meaning adding clusters significantly improves clustering quality.
The curve drops sharply at K=4, meaning adding more clusters beyond this point does not improve clustering much.
The optimal K is 4 because it balances accuracy and efficiency.

### 3. k-means clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 3) 


# Assign clusters
df_numeric['Cluster'] = kmeans.fit_predict(df_scaled)

df_numeric.head()
```

```python
# Check Cluster Distribution

df_numeric['Cluster'].value_counts()
```

```python
# Scatterplot - Cocoa Percent vs Rating (Clustered)
sns.scatterplot(x=df_numeric['Cocoa_Percent'], y=df_numeric['Rating'], hue=df_numeric['Cluster'], palette='viridis', s=50)
plt.xlabel('Cocoa Percent')
plt.ylabel('Rating')
plt.title('Clusters based on Cocoa Percent and Rating')
plt.legend(title="Cluster")
plt.show()
```

Cluster 0 (Dark Purple): High cocoa (70%-100%), mixed ratings (2.0-3.5).
Cluster 1 (Teal): Spread across cocoa levels (40%-80%), some high ratings (4.0+).
Cluster 2 (Yellow): Medium cocoa (60%-80%), consistently higher ratings (3.0-4.0).

```python
# Scatterplot - Review Date vs Rating (Clustered)
sns.scatterplot(x=df_numeric['Review_Date'], y=df_numeric['Rating'], hue=df_numeric['Cluster'], palette='viridis', s=50)
plt.xlabel('Review Date')
plt.ylabel('Rating')
plt.title('Clusters based on Review Date and Rating')
plt.legend(title="Cluster")
plt.show()
```

Cluster 0 (Dark Purple): Mostly low ratings (1.0 - 2.5) over time.
Cluster 1 (Teal): Spread across different years with moderate to high ratings (2.5 - 5.0).
Cluster 2 (Yellow): Consistently higher ratings (3.0 - 4.5+), especially in later years.

```python
cluster_summary = df_numeric.groupby('Cluster').agg({
    'Review_Date': ['mean', 'median'],
    'Cocoa_Percent': ['mean', 'median'],
    'Rating': ['mean', 'median']
})
cluster_summary
```
